## Handling Missing Values

```python
df['children'] = df['children'].fillna(0)
    
df['city'] = df['city'].fillna('Unknown')
    
df['agent'] = df['agent'].fillna(0)
    
df['company'] = df['company'].fillna(0)

df['adr'].fillna(0)

df['required_car_parking_spaces'].fillna(0)

df['total_of_special_requests'].fillna(0)

df['reservation_status'].dropna()

df.isnull().sum().sum()
```

Penjelasan
- Untuk variabel children dapat diisikan dengan nilai 0, karena banyak pengunjung yang tidak membawa anak.
- Dikarenakan terdapat 488 missing pada variabel city cukup sulit mendefinisikan nilainya, sehingga dapat diisikan dengan Unknown.
- Variabel agent dan company memiliki banyak missing value. Karena agent dan company dapat diisikan dengan nilai 0, atau dapat disimpulkan bahwa pengunjung tidak menggunakan agent atau company dalam melakukan booking.

##  Handle Duplicated Data

```python
df.duplicated().sum()
```

Terdapat 33261 data yang memiliki data duplikat.

```python
df[df.duplicated(keep=False)].head(3)
```

Data duplikat dapat di drop, karena banyak sekali variabel yang bernilai sama.

```python
df = df.drop_duplicates()

print(f"Total data: {df.shape}")
```

```python
for col in ['children', 'agent', 'company']:
    df[col] = df[col].astype(int)
```

Data yang tidak sesuai dengan tipe datanya telah dirubah, dari float menjadi integer.

```python
zero_guest = (
    (df['adults'] == 0) & 
    (df['children'] == 0) & 
    (df['babies'] == 0) &
    (df['reservation_status'] != 'Canceled')
)

zero_adult = (
    (df['adults'] == 0) & 
    (df['reservation_status'] != 'Canceled') 
)

zero_night = (
    (df['stays_in_weekend_nights'] == 0) & 
    (df['stays_in_weekdays_nights'] == 0) &
    (df['reservation_status'] != 'Canceled')
)

df1 = df.loc[(~zero_guest) | (~zero_adult)].copy()
df_cleaned = df1.loc[~zero_night].reset_index(drop=True).copy()
df_cleaned.info()
```

Data seperti zero_guest, zero_adult dan zero_night kemudian dikeluarkan sehingga data dapat diproses lebih lanjut.

```python
df_cleaned.to_csv("cleaned_hotel_bookings_data.csv", index=False)
```
